```python
import yfinance as yf
import pandas as pd

# Download historical data for AAPL from 2020-01-01 to 2023-01-01
data = yf.download("AAPL", start="2020-01-01", end="2023-01-01")

# Derive historical prices as features
data['Open_Lag1'] = data['Open'].shift(1)
data['High_Lag1'] = data['High'].shift(1)
data['Low_Lag1'] = data['Low'].shift(1)
data['Close_Lag1'] = data['Close'].shift(1)
data['Volume_Lag1'] = data['Volume'].shift(1)

# Derive technical indicators (Moving Averages)
data['MA_5'] = data['Close'].rolling(window=5).mean()
data['MA_10'] = data['Close'].rolling(window=10).mean()
data['MA_50'] = data['Close'].rolling(window=50).mean()

# Drop rows with NaN values created by shifting and rolling windows
data.dropna(inplace=True)

# Display the first few rows with new features
display(data.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Define features (X) and target (y)
# We will use the engineered features to predict the 'Close' price for the next day
X = data[['Open_Lag1', 'High_Lag1', 'Low_Lag1', 'Close_Lag1', 'Volume_Lag1', 'MA_5', 'MA_10', 'MA_50']]
y = data['Close']

# Shift the target variable to predict the next day's closing price
y = y.shift(-1)

# Drop the last row in X and y as the last y value will be NaN after shifting
X = X.iloc[:-1]
y = y.iloc[:-1]

# Split data into training and testing sets (80/20 split)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False) # shuffle=False for time series data

# Initialize and train the Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"R-squared (RÂ²): {r2:.2f}")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import numpy as np

# Create the target variable for classification
# 1 if the close price increases tomorrow, 0 otherwise
data['Price_Increase'] = (data['Close'].shift(-1) > data['Close']).astype(int)

# Define features (X) and target (y)
# Use the same features as the regression model
X = data[['Open_Lag1', 'High_Lag1', 'Low_Lag1', 'Close_Lag1', 'Volume_Lag1', 'MA_5', 'MA_10', 'MA_50']]
y = data['Price_Increase']

# Drop the last row in X and y as the last y value will be NaN after shifting
X = X.iloc[:-1]
y = y.iloc[:-1]

# Split data into training and testing sets (80/20 split)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False) # shuffle=False for time series data

# Initialize and train the Random Forest Classifier model
model = RandomForestClassifier(n_estimators=100, random_state=42) # Using 100 estimators and a random state for reproducibility
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy:.2f}")
print("Confusion Matrix:")
print(conf_matrix)
print("Classification Report:")
print(class_report)

# Report feature importance
feature_importances = pd.Series(model.feature_importances_, index=X.columns).sort_values(ascending=False)
print("\nFeature Importance:")
print(feature_importances)
```

```python
import yfinance as yf
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import numpy as np

# Download historical data for AAPL from 2024-01-01 to 2025-01-01
future_data = yf.download("AAPL", start="2024-01-01", end="2025-01-01")

# Apply the same feature engineering steps to the new data
future_data['Open_Lag1'] = future_data['Open'].shift(1)
future_data['High_Lag1'] = future_data['High'].shift(1)
future_data['Low_Lag1'] = future_data['Low'].shift(1)
future_data['Close_Lag1'] = future_data['Close'].shift(1)
future_data['Volume_Lag1'] = future_data['Volume'].shift(1)
future_data['MA_5'] = future_data['Close'].rolling(window=5).mean()
future_data['MA_10'] = future_data['Close'].rolling(window=10).mean()
future_data['MA_50'] = future_data['Close'].rolling(window=50).mean()

# Create the target variable for classification in the new data
future_data['Price_Increase'] = (future_data['Close'].shift(-1) > future_data['Close']).astype(int)

# Drop rows with NaN values created by shifting and rolling windows
future_data.dropna(inplace=True)

# Define features (X_future) and target (y_future) for the future data
X_future = future_data[['Open_Lag1', 'High_Lag1', 'Low_Lag1', 'Close_Lag1', 'Volume_Lag1', 'MA_5', 'MA_10', 'MA_50']]
y_future = future_data['Price_Increase']

# Drop the last row in X_future and y_future as the last y_future value will be NaN after shifting
X_future = X_future.iloc[:-1]
y_future = y_future.iloc[:-1]

# Ensure the columns in X_future match the columns used for training
# This is important if any columns were dropped or added during feature engineering
# In this case, the columns should be the same, but it's a good practice to ensure
X_future = X_future[X_train.columns]


# Make predictions on the future data using the trained Random Forest model
y_pred_future = model.predict(X_future)

# Evaluate the model performance on the future data
accuracy_future = accuracy_score(y_future, y_pred_future)
conf_matrix_future = confusion_matrix(y_future, y_pred_future)
class_report_future = classification_report(y_future, y_pred_future)

print(f"Accuracy on 2024-2025 data: {accuracy_future:.2f}")
print("Confusion Matrix on 2024-2025 data:")
print(conf_matrix_future)
print("Classification Report on 2024-2025 data:")
print(class_report_future)
```
